Задача: Дана функция: f(x) = 0.6x^3+5.5x^2+10x-5

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot 
x = Symbol('x')
f = 0.6*x**3 + 5.5*x**2 +10*x -5
f
```

```python
# Определение корней:
x1=solve(f)[0]
x2=solve(f)[1]
x3=solve(f)[2]
print(f"Корень x1 = {round(x1,3)}")
print(f"Корень x2 = {round(x2,3)}")
print(f"Корень x3 = {round(x3,3)}")
plot(f)
# Находим точки мин и макс:
list_x = solve(f.diff(x))
list_y = [0.6*x**3 + 5.5*x**2 +10*x -5 for x in list_x]
rez = zip(list_x,list_y)
max,min = zip(rez)
print(f'Точка max = {max}')
print(f'Точка min = {min}')
```

```python
# График функции:
import matplotlib.pyplot as plt
import numpy as np
fig , ax = plt.subplots()
x = np.linspace(-7,2,50 )
y = 0.6*x**3 + 5.5*x**2 +10*x -5
ax.plot(np.linspace(-7,2,50 ),x*0,color='black')
ax.plot(y*0,0.6*x**3 + 5.5*x**2 +10*x -5,color='black')
ax.plot(x,y)

xmin,ymin = min[0][0],min[0][1]
xmax,ymax = max[0][0],max[0][1]

ax.scatter(round(xmin,3),round(ymin,3))
ax.scatter(round(xmax,3),round(ymax,3))
ax.scatter(round(x1,3),0,)
ax.scatter(round(x2,3),0)
ax.scatter(round(x3,3),0)
plt.show()
```

```python
print(f"Функция возрастает на промежутке от -оо до {round(xmax,3),round(ymax,3)} и от {round(xmin,3),round(ymin,3)}")
print(f"Функция < 0 на промежутке от -оо до ({round(x1,3)}) и от ({round(x2,3)}) до ({round(x3,3)})")
print(f"Функция > 0 на промежутке от  ({round(x1,3)}) до ({round(x2,3)}) и от ({round(x3,3)}) до +оо")
```
