```python
def fib(n):  # Метод
    if n <= 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
# Квадратные уровнения
from math import sqrt # Можно импортировать модули или отдельные библиотеки.

def solve(a, b, c): # Метод
    d = b**2 -4*a*c # Расчет дискриминанта
    if d >= 0:
        x1 = (-b + sqrt(d)) / (2 * a) # Расчет первого корня
        x2 = (-b - sqrt(d)) / (2 * a) # Второго
        return (x1, x2)
    else:
        return 'Вещественных корней нет'
```

```python
solve(0.5, 0.125, 0)
```

```python
solve(1, 2, 3)
```

```python
import cmath # Если я импортирую модуль позволяющий вычислять, в том числе корни комплексных чисел

def csolve(a, b, c):
    d = b**2 - 4*a*c
    x1 = (-b + cmath.sqrt(d)) / (2 * a) # Расчет первого корня
    x2 = (-b - cmath.sqrt(d)) / (2 * a) # Второго
    return (x1, x2)
    
```

```python
csolve(1, 2, 5) # В ответе будет два комплексных корня.
```

```python
# Можно применять в области аналитики или проводить свои эксперементы
# Пример: Есть игральная кость. Теория вероятности говорит, что чем больше я буду опытов проделывать, тем скорее всего
# распределени будет одинаковым. Вопрос, сколько нужно опытов проделать, чтобы получить равновероятностное распределение.
```

```python
import random
import matplotlib.pyplot as plt

count = 1000 # Будет выполнено count опытов.
x = list(range(1, 7))
y = [0 for i in x]

fig, ax = plt.subplots()

# Создаю отдельный список из 6-ти нулей. Каждый из которых будет обозначать сколько раз выпала та или иная грань.
for _ in range(count):
    y[random.randint(0, 5)] += 1
    
ax.bar(x, y)

y.sort()
# print('{}%'.format(round((y[5] - y[0]) / count * 100, 3)))
stat = [f'{i + 1}: {round(y[i] / count * 100)}%' for i in range(0, 6)]
print(stat)
plt.show() # Увижу гистограмму распределения и какое значение выпало в процентном соотношении.
```
